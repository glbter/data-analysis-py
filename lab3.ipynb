### 4. Створити декілька власних об’єктів DataFrame за такою ж тематикою, що й файл. Наприклад, якщо тема файлу – жаби, можна створити об’єкти, що містять розміри жаб, вагу, стать, кількість особин в популяції і т.д. Використати описані в теоретичних відомостях параметри методу merge для різних видів злиття даних цих об’єктів.

```python
data = pd.read_csv("possum.csv")
myData1 = {'Pop' : ['Vic', 'other'],
          'Average Number Of Childs': [2, 1]}
frame1 = pd.DataFrame(myData1)
newData1 = pd.merge(data, frame1, left_on='Pop', right_on='Pop')
newData1
```

```python
myData2 = {'Pop' : ['Vic', 'KPI', 'other'],
          'Total number of possums in population': [52, 2, 323]}
frame2 = pd.DataFrame(myData2)
newData2 = pd.merge(data, frame2, left_on='Pop', right_on='Pop', how='right')
newData2
```

```python
newData2.loc[newData2['Total number of possums in population'] == 2]
```
